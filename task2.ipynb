```python
img = plt.imread("data/cells.png")

def clipping_b(image, T1, T2):

    img_clip_b = numpy.zeros(img.shape)

    for i, j in numpy.ndindex(img_clip_b.shape):

        if img[i, j] < T1:
            img_clip_b[i, j] = T1
            
        elif img[i, j] > T2:
            img_clip_b[i, j] = T2
            
        else:
            img_clip_b[i, j] = img[i, j]

    return img_clip_b


plt.imshow(clipping_b(img, T1, T2), "gray") 
plt.colorbar()

```
