**<a id="#">Task 2.3.4.</a>** Finish the implementation of the function `clip_image` below, by replacing the comment line (the one which starts with `#`) with your code:

```python
def clip_image(img, T1, T2):
    result = img #passiert hier nun alles, was ich mit "result" mache, auch mit "img" wegen des Angleichens?

    i = 0
    for i in range(0, 768):
        
        j = 0
        for j in range(0, 1024):
        
            if result[i, j] < T1:
                result[i, j] = T1
            
            elif result[i, j] > T2:
                result[i, j] = T2


    return result
```
